## q2.2

```python
import matplotlib.pyplot as plt
import numpy as np

# Speedup function
def speedup(p, N):
    return p

# Plot 1
p_values_1 = [2**i for i in range(11)]
N_1 = 1024
S_values_1 = [speedup(p, N_1) for p in p_values_1]

# Plot 2
p_2 = 8
N_values_2 = [2**i for i in range(3, 11)]
S_values_2 = [speedup(p_2, N) for N in N_values_2]

# Plot 3
p_3 = 32
N_values_3 = [2**i for i in range(5, 13)]
S_values_3 = [speedup(p_3, N) for N in N_values_3]

# Visualize
plt.figure(figsize=(15, 5))

# Subplot 1
plt.subplot(1, 3, 1)
plt.plot(p_values_1, S_values_1, '-o', label=f'N={N_1}')
plt.xlabel('p')
plt.ylabel('Speedup')
plt.title('Speedup vs Number of Processors in log2 scale (N=1024)')
plt.xscale('log', base=2)
plt.yscale('log', base=2)
plt.grid(True, which="both", ls="--")
plt.legend()

# Subplot 2
plt.subplot(1, 3, 2)
plt.plot(N_values_2, S_values_2, '-o', label=f'p={p_2}')
plt.xlabel('Size')
plt.ylabel('Speedup')
plt.title('Speedup vs Size in log2 scale (p=8)')
plt.xscale('log', base=2)
plt.grid(True, which="both", ls="--")
plt.legend()

# Subplot 3
plt.subplot(1, 3, 3)
plt.plot(N_values_3, S_values_3, '-o', label=f'p={p_3}')
plt.xlabel('Size')
plt.ylabel('Speedup')
plt.title('Speedup vs Size in log2 scale (p=32)')
plt.xscale('log', base=2)
plt.grid(True, which="both", ls="--")
plt.legend()

plt.tight_layout()
plt.show()
```
